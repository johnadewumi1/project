# DATA VISUALIZATION WITH MATPLOTLIB AND PYPLOT

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
youtube_views=[534,690,258,402,724,689,352]
days=[1,2,3,4,5,6,7]

#barchart:
plt.bar(days,youtube_views,label='Youtube Viewers',color='purple')#label signify legend
plt.xlabel('Days')
plt.ylabel('Views')
plt.title('Bar chart of youtube views')
plt.legend()
plt.savefig('youtube barchart.png',dpi=500)
plt.show()
```

```python
youtube_views=[534,690,258,402,724,689,352]
facebook_views=[123,342,700,305,406,648,325]
days=[1,2,3,4,5,6,7]

#Barchat2
plt.bar(days,youtube_views,label='Youtube Viewers')
plt.bar(days,facebook_views,label='Facebook viewers')
plt.xlabel('Days')
plt.ylabel('Views')
plt.title('social media bar chart')
plt.legend('upper left')
plt.savefig('social media barchart.png',dpi=500)
plt.show()
```

```python
#lineplot

youtube_views=[534,690,258,402,724,689,352]
facebook_views=[123,342,700,305,406,648,325]
twitter_views=[202,209,176,415,824,389,550]
days=[1,2,3,4,5,6,7]

plt.plot(days,youtube_views,label='youtube views',marker='o')
plt.plot(days,facebook_views,label='facebook views',marker='o')
plt.plot(days,twitter_views,label='youtube views',marker='o')
plt.xlabel('Days')
plt.ylabel('Views')
plt.title('social media line chart')
plt.legend()
plt.savefig('social media line chart.png',dpi=500)
plt.show()
```

```python
plt.plot(days,youtube_views,label='youtube views',marker='o')
plt.plot(days,facebook_views,label='facebook views',marker='o')
plt.plot(days,twitter_views,label='youtube views',marker='o')


plt.xlabel('Days')
plt.ylabel('Views')
plt.title('social media line chart')
plt.legend()


#plt.savefig('social media line chart.png',dpi=500)
plt.xlim(1,7.5)
plt.ylim(100,850)

plt.grid(True,color='red', linestyle='--')
plt.show()
```

```python
#Piechat
#categorical and quantitative data
social_media=['Facebook','Instagram','Youtube','Linked-In']
views=[300,350,400,450]
exp=[0,0,0,0.2]
```

```python
plt.pie(views,labels=social_media) #autopct will give the percentage, explode will removed the biggest
#plt.savefig('Social media views')
plt.show()
```

```python
plt.pie(views,labels=social_media,autopct='%1.1f%%',explode=exp,shadow=True) #autopct will give the percentage, explode will removed the biggest
#plt.savefig('Social media views')
plt.show()
```

```python
#HISTOGRAM
ages=[22,55,62,45,21,22,99,34,42,4,102,110,27,48,100,84]
bins=[0,20,40,60,80,100]

plt.hist(ages,bins=bins,color='Purple')
plt.xlabel('Ages')
plt.ylabel('Frequency')
plt.title('A chart of Ages')
#plt.savefig()
plt.show()
```

```python
#SCATTERED PLOT
#only done for quantitave variablesangaist another quantitative
youtube_views=[534,690,258,402,724,689,352]
facebook_views=[123,342,700,305,406,648,325]
twitter_views=[202,209,176,415,824,389,550]
days=[1,2,3,4,5,6,7]

plt.scatter(days,youtube_views,label='Youtubeviewers', marker='o')
plt.scatter(days,facebook_views,label='Facebookviewers', marker='o')
plt.scatter(days,twitter_views,label='twittervierwer', marker='o')

plt.xlabel('Days')
plt.ylabel('Views')

plt.title('Social Media Views')
plt.legend()

plt.xlim(0,7.5)
plt.ylim(100,830)

plt.show()
```

```python
#catplot

tips_data=sns.load_dataset('tips')
tips_data
```

```python
#getting the day that has cstomer
sns.catplot(x='day',y='tip',data=tips_data,kind='bar')
plt.show()
```

```python
#who pay more tips
sns.catplot(x='sex',y='tip',data=tips_data,kind='violin')
plt.show()
```
