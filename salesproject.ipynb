```python
def Ageband(x):
    if x>=30 and x<45:
        return '30-44'
    elif  x>=45 and x<60:
        return '45-59'
    elif  x>=60 and x<75:
        return '60-74'
    elif  x>=75 and x<90:
        return '75-89'
    elif  x>=90 and x<105:
        return '90-104'
    elif x>=105 and x<120:
        return '105-119'
    else:
        return '0'
```

## Condtioning the shipping cost to derive cost level

```python
avgcost=12.84
def average(x):
    if x>avgcost:
        return 'High Cost'
    elif x==avgcost:
        return 'Average Cost'
    else:
        return 'Low Cost'
```

```python
Cost_level=sales_d['Shipping Cost'].apply(average)
Cost_level
```

```python
sales_d.insert(21,'CostLevel',list(Cost_level),True)
```
